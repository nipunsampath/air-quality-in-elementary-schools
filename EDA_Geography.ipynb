```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme()
sns.set_style("whitegrid")

base_path = f"{os.environ["OneDrive"]}/Data Science Project Idea 31"
```

```python
# Loading data
geography = pd.read_csv(f"{base_path}/geography.csv")
geography.head()
```

```python
# Measures of central tendencies
geography.describe()
```

```python
geography.isnull().sum()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[12, 8])
fig.suptitle("Histograms of Demographics")

geography["Distance_to_freeway"].hist(bins=5, ax=axes[0, 0])
axes[0, 0].set_title("Distance_to_freeway")
geography["Street_100"].hist(bins=5, ax=axes[0, 1])
axes[0, 1].set_title("Street_100")
geography["Street_200"].hist(bins=5, ax=axes[0, 2])
axes[0, 2].set_title("Street_200")
geography["Street_300m"].hist(bins=5, ax=axes[1, 0])
axes[1, 0].set_title("Street_300m")
geography["PercentTree"].hist(bins=5, ax=axes[1, 1])
axes[1, 1].set_title("PercentTree")

fig.tight_layout()
```

```python
# correlation
geography_corr = geography.corr()
sns.heatmap(geography_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.scatterplot(data=geography, x='Distance_to_freeway', y='PercentTree')
plt.show()
```

```python
geography[['Distance_to_freeway', 'Street_100', 'Street_200', 'Street_300m', 'PercentTree']].boxplot()

plt.title('Box Plots of Geography')
plt.tight_layout()
plt.show()
```
